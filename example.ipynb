### Input geojson of area of interest here.

### Run the geojson_to_lia_rasters_and_lia_stack() utility. Depending on how large your AOI is, this may take a moment.

```python
generate_lia.geojson_to_lia_rasters_and_lia_stack(aoi_geojson)
```

### Let's check out the netcdf stack to view the local incidence angle maps per relative orbit. For visualization purposes, let's convert to degrees and mask out angles less than 15 and greater than 75.

```python
lia_stack = xr.open_dataarray('lia_maps/lia_stack_orbits_13_64_115_137.nc',decode_coords='all')
lia_stack_degrees = np.rad2deg(lia_stack)
lia_stack_degrees = lia_stack_degrees.where((lia_stack_degrees>15) & (lia_stack_degrees<75))
lia_stack_degrees.plot(col='rel_orbit')
```
